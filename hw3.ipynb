```python
import numpy as np
redo =  np.array([[-np.inf, 2, 3], [-np.inf, -np.inf, 4], [1, 5, 6]])
newDim = (redo.shape[0]*redo.shape[1])
redo = list(redo.reshape(newDim))
print(redo)
total = [0,0]
for i in range(9):
  if (redo[i] != -np.inf): total = [sum(value) for value in zip(total, [1,redo[i]] )]
redo = np.array(list(map(lambda x: total[1]/total[0] if(x == -np.inf) else x, redo))).reshape([3,3])
redo
```

```python
import pandas as pd

students = pd.DataFrame(
 {
 "id": range(101, 111),
 "courses": [
 ["A", "B", "C"],
 ["A", "B", "E"],
 ["B", "D", "E"],
 ["C", "D", "F"],
 ["A", "D", "F"],
 ["B", "C", "E"],
 ["A", "D", "E"],
 ["C", "E", "F"],
 ["D", "E", "F"],
 ["B", "E", "F"],
 ],
 }
)
long_list = []
for array in students["courses"].to_numpy():
  long_list += array
freq = {}
for item in long_list:
    if (item in freq):
        freq[item] += 1
    else:
        freq[item] = 1
print(freq)
freq = pd.DataFrame([freq['A'],freq['B'],freq['C'],freq['D'],freq['E'], freq['F']], index = ['A','B','C','D','E','F']).sort_values(by=[0],ascending = False)
freq
```
